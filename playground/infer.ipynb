# Load Model

```python
from modelscope.models import Model
from modelscope.pipelines import pipeline
from modelscope.utils.constant import Tasks
from modelscope.preprocessors import TokenClassificationTransformersPreprocessor
```

```python
model_id = 'iic/nlp_structbert_word-segmentation_chinese-base'
model = Model.from_pretrained(model_id)
tokenizer = TokenClassificationTransformersPreprocessor(model.model_dir, max_length=512)
pipeline_ins = pipeline(task=Tasks.word_segmentation, model=model, preprocessor=tokenizer)
```

# icwb2-data

```python
def load_data(path, split=False):
    with open(path, 'r', encoding='utf-8') as f:
        if not split:
            return [line.rstrip('\n') for line in f.readlines() if len(line.rstrip('\n')) < 510]
        else:
            return [line.split() for line in f.readlines() if len("".join(line.split())) < 510]
```

```python
dname = "pku"
data_path = f"D:\\Local\\Workspace\\CSE256\\AdaSeq\\playground\\datasets\\icwb2-data\\testing\\{dname}_test.utf8"
target_path = f"D:\\Local\\Workspace\\CSE256\\AdaSeq\\playground\\datasets\\icwb2-data\\gold\\{dname}_test_gold.utf8"

data = load_data(data_path)
target = load_data(target_path, split=True)

print(len(data))
for d in data[:10]:
    print(d)

print(len(target))
for t in target[:10]:
    print(t)
```

```python
result = pipeline_ins(data, batch_size=128)
output = [o["output"] for o in result]
print(len(output))
for o in output[:10]:
    print(o)
```

```python
import json

output_path = f"D:\\Local\\Workspace\\CSE256\\AdaSeq\\playground\\outputs\\{dname}_output.json"
target_path = f"D:\\Local\\Workspace\\CSE256\\AdaSeq\\playground\\outputs\\{dname}_target.json"
with open(output_path, "w", encoding="utf-8") as file:
    json.dump(output, file, ensure_ascii=False)
with open(target_path, "w", encoding="utf-8") as file:
    json.dump(target, file, ensure_ascii=False)
```

# NLPCC2018-test

```python
import csv

def load_data(path):
    data = []
    target = []
    with open(path, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        for i, line in enumerate(reader):
            if i == 0:
                continue
            s = line[0].rstrip('\n')
            t = s.split()
            d = "".join(t)
            if len(d) < 510:
                data.append(d)
                target.append(t)
    return data, target
```

```python
data_path = "D:\\Local\\Workspace\\CSE256\\AdaSeq\\playground\\datasets\\NLPCC2018-test\\nlpcc_2018_test.csv"
data, target = load_data(data_path)

print(len(data))
for d in data[:10]:
    print(d)

print(len(target))
for t in target[:10]:
    print(t)
```

```python
result = pipeline_ins(data, batch_size=128)
output = [o["output"] for o in result]
print(len(output))
for o in output[:10]:
    print(o)
```

```python
import json

output_path = f"D:\\Local\\Workspace\\CSE256\\AdaSeq\\playground\\outputs\\nlpcc_output.json"
target_path = f"D:\\Local\\Workspace\\CSE256\\AdaSeq\\playground\\outputs\\nlpcc_target.json"
with open(output_path, "w", encoding="utf-8") as file:
    json.dump(output, file, ensure_ascii=False)
with open(target_path, "w", encoding="utf-8") as file:
    json.dump(target, file, ensure_ascii=False)
```
